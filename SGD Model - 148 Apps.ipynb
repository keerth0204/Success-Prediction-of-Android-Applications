```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, KFold
```

```python
df = pd.read_excel('AppsWithSenti.xlsx')
```

```python
df.head()
```

```python
df = df.drop(['Unnamed: 0', 'Reviews', 'Lemm_sentiment'], axis=1)
```

```python
df.head()
```

```python
X = df.drop('Overall', axis=1)
y = df.Overall
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled = pd.DataFrame(X_scaled)
```

```python
X_scaled.columns = ['Graphics', 'UI', 'Gameplay', 'Replay Value', 'Sentiment']
```

```python
X_scaled.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y)
```

```python
from sklearn.linear_model import SGDRegressor
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
```

```python
kfold = KFold(n_splits=10)
cv_results = cross_val_score(SGDRegressor(), X_train, y_train, 
                             cv=kfold, scoring='neg_mean_squared_error')
print(cv_results.mean())
```

```python
sgd = SGDRegressor()
sgd.fit(X_train, y_train)
pred = sgd.predict(X_test)
print(mean_squared_error(y_test, pred))
```
